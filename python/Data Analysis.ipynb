```python
import numpy as np
import matplotlib.pyplot as plt
import h5py
```

```python
plt.style.use('prl')
```

```python
df = h5py.File("scratch/sfmc_test_run/snapshots/snapshots_s20.h5","r")
```

```python
plt.subplot(121)
plt.imshow(df['tasks/psi'][0])
plt.title(r"$\psi$")
plt.subplot(122)
plt.imshow(df['tasks/phi'][0])
plt.title(r"$\phi$")
```

```python
with h5py.File("scratch/sfmc_test_run/timeseries/timeseries_s1.h5","r") as ts:
    nusselt = ts['tasks/Nusselt'][:,0,0]
    tau_phi = ts['tasks/tau_phi'][:,0,0]
    tau_psi = ts['tasks/tau_psi'][:,0,0]
    time = ts['tasks/Nusselt'].dims[0][0][:]
```

```python
plt.semilogy(time,nusselt,'x-')
plt.xlabel("time")
plt.ylabel("Nusselt")
```

```python
plt.plot(time,tau_phi,'x-', label=r'$\tau_\phi$')
plt.plot(time,tau_psi,'o-', label=r'$\tau_\psi$')
plt.legend()
plt.xlabel("time")
plt.ylabel(r"$\tau$")
```
